```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np

# Preprocesado y modelado
# ==============================================================================
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Configuraci√≥n warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
#Cargamos los datos!
url = 'https://raw.githubusercontent.com/JoaquinAmatRodrigo/' \
       + 'Estadistica-machine-learning-python/master/data/ESL.mixture.csv'
datos = pd.read_csv(url)
datos.head()
```

```python
# Divisi√≥n de los datos en train y test
X = datos.drop(columns = 'y')
y = datos['y']

X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y.values.reshape(-1,1),
                                        train_size   = 0.7,
                                        random_state = 1234,
                                        shuffle      = True
                                    )
```

```python
# Creaci√≥n del modelo SVM 
modelo = SVC(C = 100, kernel = 'linear', random_state=123)
modelo.fit(X_train, y_train)
```

```python
#Predicciones!
y_train_pred = modelo.predict(X_train)
y_test_pred = modelo.predict(X_test)
```

```python
#Calculo el accuracy en Train
train_accuracy = accuracy_score(y_train, y_train_pred)

print('% de aciertos sobre el set de evaluaci√≥n:',train_accuracy)
```

```python
#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% de aciertos sobre el set de evaluaci√≥n:',test_accuracy)
```
