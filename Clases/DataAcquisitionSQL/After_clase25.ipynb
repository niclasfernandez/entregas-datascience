```python
def describe_object(df, name_column):
    print('Column: ' + name_column)
    print('Unique Values: ' )
    print(df[name_column].unique())
    print('Frequency: ' )
    print(df[name_column].value_counts(normalize=True))
    
describe_object(df_housing, 'MSZoning')
```

```python
describe_object(df_housing, 'MasVnrType')
```

```python
number_df = df_housing.select_dtypes(include='number')
```

```python
col_number = number_df.columns
```

```python
def describe_number(df, name_column):
    print('Column: ' + name_column)
    print('Unique Values: ' )
    print(df[name_column].describe())
    
    
describe_number(df_housing, 'SalePrice')
```

```python
for column in col_number:
    describe_number(df_housing, column)
```

```python
import altair as alt
```

```python
sample = df_excel.sample(n = 500, random_state=8)
```

```python
base = alt.Chart(sample)
```

```python
chart = base.mark_circle()
```

```python
chart.encode(x='Quantity', y='UnitPrice')
```

```python
alt.Chart(sample).mark_circle().encode(x='Quantity', y='UnitPrice', color='Country')
```

```python
alt.Chart(sample).mark_bar().encode(alt.X("Quantity:Q", bin=alt.Bin(step=10)), \
y='count()')
```
