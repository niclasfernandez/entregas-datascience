```python
def print_faces(images, target, top_n):
    # set up the figure size in inches
    fig = plt.figure(figsize=(10, 10))
    fig.subplots_adjust(left=0, right=1, bottom=0,top=1,
                        hspace=0.05, wspace=0.05)
    for i in range(top_n):
        # plot the images in a matriz of 20X20
        p = fig.add_subplot(5, 5, i+1, xticks=[],
                            yticks=[])
        p.imshow(images[i], cmap=plt.cm.bone)
        
        # label the image with the target value
        p.text(0, 14, str(target[i]), fontsize=25)
        p.text(0, 60, str(i), fontsize=25)    
```

```python
print_faces(faces.images, faces.target, 20)
```

## Entrenar el model SVM

```python
from sklearn.svm import SVC
```

```python
svc_1 = SVC(kernel='linear')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(faces.data, faces.target, test_size=0.25, random_state=0)
```

```python
faces.data
```

```python
from sklearn import metrics
```

```python
svc_1.fit(X_train, y_train)
print('Accuarcy on training set:')
print(svc_1.score(X_train, y_train))
print('Accuarcy on test set:')
print(svc_1.score(X_test, y_test))
y_pred = svc_1.predict(X_test)

print('Classification Report:')
print(metrics.classification_report(y_test,y_pred))
print('Confusion Matrix:')
print(metrics.confusion_matrix(y_test,y_pred))
```

```python
eval_faces = [np.reshape(a, (64,64)) for a in X_test]
print_faces(eval_faces, y_pred, 10)
```

```python
glasses = [(10, 19), (30, 32), (37, 38), (50, 59), (63, 64),
(69, 69), (120, 121), (124, 129), (130, 139), (160,
161),
(164, 169), (180, 182), (185, 185), (189, 189),
(190, 192),
(194, 194), (196, 199), (260, 269), (270, 279),
(300, 309),
(330, 339), (358, 359), (360, 369)]
```

```python
def create_target(segments):
    # create a new y array of target size initialized with zeros
    y = np.zeros(faces.target.shape[0])
    # put 1 in the specified segments
    for (start, end) in segments:
        y[start:end + 1] = 1
    return y
target_glasses = create_target(glasses)
```

```python
X_train, X_test, y_train, y_test = train_test_split(faces.data, target_glasses, test_size=0.25, random_state=0)
```

```python
def train_and_evaluate(clf, X_train,X_test, y_train,y_test):
    clf.fit(X_train, y_train)
    print('Accuarcy on training set:')
    print(clf.score(X_train, y_train))
    print('Accuarcy on test set:')
    print(clf.score(X_test, y_test))
    y_pred = clf.predict(X_test)

    print('Classification Report:')
    print(metrics.classification_report(y_test,y_pred))
    print('Confusion Matrix:')
    print(metrics.confusion_matrix(y_test,y_pred))
```

```python
svc_2 = SVC(kernel='linear')
train_and_evaluate(svc_2, X_train, X_test, y_train, y_test)
```

```python
svc_2.fit(X_train, y_train)
y_pred = svc_2.predict(X_test)
y_pred
```

```python
eval_faces = [np.reshape(a, (64,64)) for a in X_test]
print_faces(eval_faces, y_pred, 10)
```
