Para nuestro primer ejemplo usaremos el famoso dataset de Iris!

```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('./IRIS.csv')
```

```python
#Como siempre realizaremos un peque√±o EDA del dataset
df.head()
```

```python
df.shape
```

```python
#FacetGrid - sepal_length vs sepal_width
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.FacetGrid(df,hue = 'species' , height = 5, aspect=2).map(plt.scatter,'sepal_length','sepal_width').add_legend();
    plt.show()
```

```python
#FacetGrid - sepal_length vs petal_length
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.FacetGrid(df,hue = 'species' , height = 5, aspect=2).map(plt.scatter,'sepal_length','petal_length').add_legend();
    plt.show()
```

```python
#FacetGrid - sepal_length vs petal_width
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.FacetGrid(df,hue = 'species' , height = 5).map(plt.scatter,'sepal_length','petal_width').add_legend();
    plt.show()
```

```python
#FacetGrid - sepal_width vs petal_length
sns.FacetGrid(df,hue = 'species' , size = 5).map(plt.scatter,'sepal_width','petal_length').add_legend();
plt.show()
#Completar
```

```python
#FacetGrid  - sepal_width vs petal_width
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.FacetGrid(df,hue = 'species' , height = 5).map(plt.scatter,'sepal_width','petal_width').add_legend();
    plt.show()
#Completar
```

```python
#FacetGrid - petal_length vs petal_width
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.FacetGrid(df,hue = 'species' , height = 5).map(plt.scatter,'petal_length','petal_width').add_legend();
    plt.show()
#Completar
```

```python
#Pairplot
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.pairplot(df,hue = 'species' , height = 4 , palette = 'Set2')
```
