# Introducci贸n

La clasificaci贸n es una t茅cnica de aprendizaje supervisado que implica predecir la etiqueta de clase para los datos de entrada dados. Al evaluar el modelo, podemos medir la calidad de nuestro modelo y podemos ver qu茅 tan bien puede funcionar nuestro modelo con respecto a nuestro caso de uso.

Las m茅tricas m谩s habituales son las siguientes: 
* Exactitud
* Matriz de confusi贸n

```python
# Importamos las librerias
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Carga del dataset
cancer_data = load_breast_cancer()

# Separamos en X e y
X = cancer_data.data
y = cancer_data.target

# Training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
print(f"Train Data: {X_train.shape}, {y_train.shape}")
print(f"Train Data: {X_test.shape}, {y_test.shape}")

# Creamos un Random Forest y lo entrenamos!
classifier = RandomForestClassifier(random_state=18)
classifier.fit(X_train, y_train)

# Calculamos las predicciones!
predictions = classifier.predict(X_test)
```

Vamos a iniciar calculando el "Accuracy", recordar que este m茅trica no es la m谩s adecuada a utilizar cuando tenemos dataset desbalanceados.



```python
from sklearn.metrics import accuracy_score

# Accuracy
print(f"Accuracy of the classifier is: {accuracy_score(y_test, predictions)}")
```

Ahora vamos a calcular la Matriz de Confusi贸n

**Recordemos**

Hay cuatro t茅rminos importantes en una matriz de confusi贸n:

* Verdaderos positivos (TP): estos son los casos en los que el "S铆" predicho en realidad pertenec铆a a la clase "S铆".
* Negativos verdaderos (TN): Estos son los casos en los que el No predicho en realidad pertenec铆a a la clase No.
* Falsos positivos (FP): estos son los casos en los que el "S铆" predicho en realidad pertenec铆a a la clase "No".
* Falsos negativos (FN): estos son los casos en los que el "No" predicho en realidad pertenec铆a a la clase "S铆".

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

#Matriz de Confusi贸n
cm = confusion_matrix(y_test, predictions)
print(cm)

#Ploteamos la Matriz
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

plt.show()
```

La precisi贸n es la proporci贸n de verdaderos positivos (TP) por la suma de verdaderos positivos (TP) y falsos positivos (FP).



Para ilustrar esta m茅trica, tomemos un ejemplo de caso de uso de detecci贸n de spam. Si nuestro modelo detecta un correo como correo no deseado que en realidad no era un correo no deseado, el usuario podr铆a perder un correo importante, es decir, aqu铆 se deber铆an reducir los falsos positivos. Entonces, en este caso de uso, debemos usar la precisi贸n como m茅trica para medir la calidad de nuestro clasificador.

```python
from sklearn.metrics import precision_score

# Precision!
print(f"Precision Score of the classifier is: {precision_score(y_test, predictions)}")
```

El Recall es la proporci贸n de verdaderos positivos (TP) por la suma de verdaderos positivos (TP) y falsos negativos (FN).



 Tomemos un ejemplo de caso de uso de predicci贸n de c谩ncer. Considere una persona que en realidad tiene c谩ncer, pero nuestro clasificador predijo que no era un paciente con c谩ncer, lo que puede conducir a un maltrato de la persona, es decir, aqu铆 se deben reducir los falsos negativos. Entonces, en este caso, debemos usar la recuperaci贸n como m茅trica para medir la calidad de nuestro clasificador.

```python
from sklearn.metrics import recall_score

# Recall
print(f"Recall Score of the classifier is: {recall_score(y_test, predictions)}")
```

La puntuaci贸n F1 es la media arm贸nica de precisi贸n y el recall. Se encuentra entre [0,1].




```python
from sklearn.metrics import f1_score

# Calculo del F1 score
print(f"F1 Score of the classifier is: {f1_score(y_test, predictions)}")
```

**Curva ROC** 

La curva AUC-ROC es una m茅trica de rendimiento que se utiliza para medir el rendimiento del modelo de clasificaci贸n en diferentes valores de umbral. Cuanto mayor sea el valor de AUC (rea bajo la curva), mejor ser谩 nuestro clasificador para predecir las clases. AUC-ROC se utiliza principalmente en problemas de clasificaci贸n binaria.

La curva ROC se traza entre la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR), es decir, TPR en el eje y y FPR en el eje y. AUC es el 谩rea bajo la curva ROC. 

Un clasificador excelente tiene un valor AUC cercano a 1, mientras que un clasificador de bajo rendimiento tiene un valor AOC cercano a 0. Un clasificador con una puntuaci贸n AOC de 0.5 no tiene ninguna capacidad de separaci贸n de clases.



```python
#Template CURVA - ROC

from sklearn.metrics import roc_curve, auc

class_probabilities = classifier.predict_proba(X_test)
preds = class_probabilities[:, 1]

fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)

# AUC
print(f"AUC for our classifier is: {roc_auc}")

# Gr谩fica de la Curva ROC
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

## Extra

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

pipe_lr = make_pipeline(StandardScaler(),
                        PCA(n_components=2),
                        LogisticRegression())

pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
test_acc = pipe_lr.score(X_test, y_test)
print(f'Test accuracy: {test_acc:.3f}')
```

## Optimizing the precision and recall of a classification model

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import matthews_corrcoef

pre_val = precision_score(y_true=y_test, y_pred=y_pred)
print(f'Precision: {pre_val:.3f}')

rec_val = recall_score(y_true=y_test, y_pred=y_pred)
print(f'Recall: {rec_val:.3f}')

f1_val = f1_score(y_true=y_test, y_pred=y_pred)
print(f'F1: {f1_val:.3f}')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import make_scorer


pipe_svc = make_pipeline(StandardScaler(),
                         SVC(random_state=1))


scorer = make_scorer(f1_score, pos_label=0)

c_gamma_range = [0.01, 0.1, 1.0, 10.0]


param_grid = [{'svc__C': c_gamma_range,
               'svc__kernel': ['linear']},
              {'svc__C': c_gamma_range,
               'svc__gamma': c_gamma_range,
               'svc__kernel': ['rbf']}]

gs = GridSearchCV(estimator=pipe_svc,
                  param_grid=param_grid,
                  scoring=scorer,
                  cv=10,
                  n_jobs=-1)
gs = gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
```
