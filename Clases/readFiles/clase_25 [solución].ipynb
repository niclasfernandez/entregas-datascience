```python
df = pd.read_csv ('bank-additional-full.csv', sep = ';')
df.head()
```

```python
df = pd.read_csv ('bank-additional-full.csv', nrows= 200)
df.shape
```

```python
meses = ['may', 'jul']
df = pd.DataFrame()
for trozo in pd.read_csv('bank-additional-full.csv', sep=';',
                             chunksize=1000):
    df = pd.concat([df,trozo[trozo['month'].isin(meses)]])

df.month.value_counts()
```

```python
df2 = pd.DataFrame()
for trozo in pd.read_csv('bank-additional-full.csv', sep=';',
                             chunksize=1000):
    df2 = pd.concat([df2,trozo.sample(frac=0.25)])
df2.shape
```
