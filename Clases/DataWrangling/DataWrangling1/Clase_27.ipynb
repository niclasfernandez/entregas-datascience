El cÃ³lico es una condiciÃ³n dolorosa que pueden sufrir los caballos, y este conjunto de datos contiene varios datos relacionados con casos especÃ­ficos de esta condiciÃ³n.

```python
file = 'dataset/horse-colic.data'
df_colic = pd.read_csv(file, header=None, sep='\s+', prefix='X')

#columns_name = [print('X'+ str(x)) for x in range(0, len(df_colic.columns))]
```

```python
df_colic = pd.read_csv(file, header=None, sep='\s+', prefix='X', na_values='?')
```

```python
df_colic.head()
```

```python
df_colic.dtypes
```

```python
df_colic.isna().sum()
```

```python
x0_mask = df_colic['X0'].isna()
```

```python
df_colic['X0'].isna().sum()
```

Crea **ciclo for** que recorrerÃ¡ todas las columnas del DataFrame. En el ciclo for, calcula la mediana para cada columna y guÃ¡rdelo en una variable llamada **col_median**. Luego, imputa los valores que faltan con este valor medio usando el mÃ©todo **.fillna()**, junto con el parÃ¡metro **inplace=True**, e imprime el nombre de la columna y su valor medio:

```python
for col_name in df_colic.columns:
    col_median = df_colic[col_name].median()
    df_colic[col_name].fillna(col_median, inplace=True)
    print(col_name)
    print(col_median)
```

```python
df_colic.isna().sum()
```

*PrÃ¡ctiquemos*


La mayorÃ­a de las organizaciones almacenan sus datos en bases de datos, almacenes de datos o los llamados data lakes. El flujo de informaciÃ³n puede provenir de diferentes sistemas o herramientas. La mayorÃ­a de las veces, los datos se almacenan en una base de datos relacional compuesto por varias tablas en lugar de una sola con relaciones bien definidas entre ellas.

Por ejemplo, una tienda en lÃ­nea podrÃ­a tener varias tablas para registrar todas las compras realizadas en su plataforma. Una tabla puede contener informaciÃ³n relacionada con los clientes existentes, otra puede enumerar todos los productos existentes y pasados en el catÃ¡logo, y una tercera puede contener todas las transacciones que ocurrieron, y asÃ­ sucesivamente.

Si estaba trabajando en un proyecto recomendando productos a los clientes para una plataforma de comercio electrÃ³nico como Amazon, es posible que solo le hayan proporcionado los datos
de la tabla de transacciones. En ese caso, le gustarÃ­a obtener algunos atributos para cada producto y cliente y tendrÃ­a que solicitar extraer estas tablas adicionales que necesita y luego fusionar las tres tablas antes de construir su sistema de recomendaciÃ³n.

Veamos cÃ³mo podemos fusionar varias fuentes de datos con un ejemplo real: el conjunto de datos de comercio minorista en lÃ­nea que usamos previamente (Online Retail). Agregaremos nueva informaciÃ³n sobre si las transacciones se realizaron en dÃ­as festivos en el Reino Unido o no. Estos datos adicionales pueden ayudar al modelo a comprender si existen algunas correlaciones entre las ventas y algunos dÃ­as festivos como la Navidad.

```python
file = 'dataset/Online Retail.xlsx'
df_retail = pd.read_excel(file)
```

```python
df_retail.head()
```

```python
uk_holidays_2022 = pd.read_csv('https://date.nager.at/PublicHoliday/Country/GB/2011/CSV')
```

```python
uk_holidays_2022.head()
```

```python
uk_holidays_2022.shape
```

```python
uk_holidays_2021 = pd.read_csv('https://date.nager.at/PublicHoliday/Country/GB/2010/CSV')
```

```python
uk_holidays_2021.shape
```

```python
uk_holidays = uk_holidays_2022.append(uk_holidays_2021)
uk_holidays.shape
```

```python
df_retail['InvoiceDate'].head()
```

```python
uk_holidays['Date'].head()
```

```python
# we need to transform the InvoiceDate column into date format (yyyy-mm-dd

df_retail['InvoiceDay'] = df_retail['InvoiceDate'].astype(str)\
                                                    .str.slice(stop=10)
df_retail.head()
```

**Left Join**

ConsevarÃ¡s todos los renglones del primer DF (Online Retail) y lo unirÃ¡s con las filas correspondientes del segundo DF (UK Public Holidays).

```python
df_left = pd.merge(df_retail, uk_holidays, left_on='InvoiceDay', right_on='Date', how='left')
df_left.shape
```

```python
df_left.head()
```

**Right Join**

ConsevarÃ¡s todos los renglones del segundo DF (UK Public Holidays) y lo unirÃ¡s con las filas correspondientes del segundo DF (Online Retail).

```python
df_right = df_retail.merge(uk_holidays, left_on='InvoiceDay', right_on='Date', how='right')
df_right.shape
```

```python
df_right.head()
```

**Inner Join**
Converva los elementos que estÃ¡n en ambos DF

```python
df_inner = df_retail.merge(uk_holidays, left_on='InvoiceDay', \
right_on='Date', how='inner')
df_inner.shape
```

**Outer Join**

Conserva todos los elementons en ambos DF

```python
df_outer = df_retail.merge(uk_holidays, left_on='InvoiceDay', \
right_on='Date', how='outer')
df_outer.shape
```
