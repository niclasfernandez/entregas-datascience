## Unas palaras sobre los métodos que devuelven una copia en vez de modificar el objeto
Parece rebuscadísimo, pero es bastante simple. Cuando nosotros llamamos a un método para que opere sobre un objeto, puede estar modificándolo o no. En caso de pandas, los métodos suelen por defecto retornar una copia del DataFrame en lugar de modificar la estructura original de los datos.
Esto nos ayuda mucho a la hora de hacer análisis exploratorio y probar cosas, ya que siempre vamos a poder volver al estado anterior de manera sencilla. Es por esto que conviene ir operando y almacenando con nombres claros las diferentes cosas que hagamos. Veamos unos ejemplos de esto

```python
#Metodos que modifican o no
lista=[1,2,3,4,5]
serie=pd.Series([1,2,3,4,5])

print(lista)
print(serie)
```

Acá tenemos dos objetos: uno es una lista y otro es una serie. Ambos son mutables así que, por ejemplo, podemos remover algún valor. Veamos cómo se comporta

```python
#Borramos el valor 5
lista.remove(5)
```

```python
#Borramos el valor 5 al darle drop al índice 4
serie.drop(4)
```

```python
#Chequeemos
print(lista)
print(serie)
```

Qué pasó acá? .drop() nos retorna <b>una copia</b> de la serie (o dataframe). Mientras que .remove() modifica a nuestro objeto original. Lo que significa que podemos hacer esto de nuevo

```python
serie.drop(4)
```

Pero no esto

```python
try:
    lista.remove(5)
except ValueError as e:
    print(e)
```

Cómo podemos hacer para que se guarden nuestros cambios entonces? Hay dos formas

```python
#1) Asignar el resultado a una variable

serie_sin_5 = serie.drop(4)
serie_sin_5
```

```python
#2) Forzar el cambio a través del parámetro inplace
serie.drop(4, inplace=True)
```

```python
serie
```

Toda vez que vean que un método o función en pandas tiene la opción <b>inplace</b> pueden optar por darle True para que modifique el objeto.
Personalmente, creo que no es lo recomendable porque <u>perdemos el estado anterior</u> y solo podremos recuperarlo volviendo a traer todo de nuevo... algo poco deseable si trabajamos con bases de datos muy grandes

```python
#Veremos que  no podemos volver a darle drop al indice 4 porque ya no existe
try:
    serie.drop(4)
except KeyError as e:
    print(e)
```

```python
serie=pd.Series([1,2,3,4,5])

#Es más recomendable tener diferentes estados de la base cruda para trabajar, tal como se hizo en el resto del notebook
serie_sin_pares= serie[serie %2 != 0]
serie_sin_impares= serie[serie %2 == 0]
serie_al_cuadrado= serie**2

print(serie_sin_pares,
      serie_sin_impares,
      serie_al_cuadrado,
      serie,
      
      sep='\n\n')

#Como verán, nuestra serie original quedó sin tocar pero pudimos obtener diferentes objetos para trabajar
```
