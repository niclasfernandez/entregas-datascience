```python
from jupyter_dash import JupyterDash     # pip install dash           
import dash_html_components as html
import dash_core_components as dcc   # Core components
from dash.dependencies import Output, Input

from dash_extensions import Lottie       # pip install dash-extensions (Lottie)
import dash_bootstrap_components as dbc  # pip install dash-bootstrap-components
import plotly.express as px              # pip install plotly
import pandas as pd                      # pip install pandas
from datetime import date
import calendar
from wordcloud import WordCloud          # pip install wordcloud

# Lottie by Emil - https://github.com/thedirtyfew/dash-extensions
url_coonections = "https://assets3.lottiefiles.com/packages/lf20_2LdLki.json" # Definir icono 1
url_companies = "https://assets9.lottiefiles.com/packages/lf20_EzPrWM.json" # icono 2
url_msg_in = "https://assets9.lottiefiles.com/packages/lf20_8wREpI.json"
url_msg_out = "https://assets2.lottiefiles.com/packages/lf20_Cc8Bpg.json"
url_reactions = "https://assets2.lottiefiles.com/packages/lf20_nKwET0.json"
options = dict(loop=True, autoplay=True, rendererSettings=dict(preserveAspectRatio='xMidYMid slice'))

# Import los datos desde CSV **************************************
df_cnt = pd.read_csv("https://raw.githubusercontent.com/Coding-with-Adam/Dash-by-Plotly/master/Analytic_Web_Apps/Linkedin_Analysis/Connections.csv")
df_cnt["Connected On"] = pd.to_datetime(df_cnt["Connected On"])
df_cnt["month"] = df_cnt["Connected On"].dt.month
df_cnt['month'] = df_cnt['month'].apply(lambda x: calendar.month_abbr[x])

df_invite = pd.read_csv("https://raw.githubusercontent.com/Coding-with-Adam/Dash-by-Plotly/master/Analytic_Web_Apps/Linkedin_Analysis/Invitations.csv")
df_invite["Sent At"] = pd.to_datetime(df_invite["Sent At"])

df_react = pd.read_csv("https://raw.githubusercontent.com/Coding-with-Adam/Dash-by-Plotly/master/Analytic_Web_Apps/Linkedin_Analysis/Reactions.csv")
df_react["Date"] = pd.to_datetime(df_react["Date"])

df_msg = pd.read_csv("https://raw.githubusercontent.com/Coding-with-Adam/Dash-by-Plotly/master/Analytic_Web_Apps/Linkedin_Analysis/messages.csv")
df_msg["DATE"] = pd.to_datetime(df_msg["DATE"])


##### CREAR EL LAYOUT ###############
# Bootstrap themes: https://hellodash.pythonanywhere.com/theme_explorer
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.COSMO]) # Aqui definimos el formato del dashboard

app.layout = dbc.Container([ # Abrir el container ( es como la estructura o cascaron de la app)
    dbc.Row([ # Abrir primera fila  (1) 
        dbc.Col([ # Columna 1 fila 1
            dbc.Card([ # Mas acerca de objetos : https://dash-bootstrap-components.opensource.faculty.ai/docs/components/card/
                dbc.CardImg(src='/assets/linkedin-logo2.png') # Aqui se debe tener la imagen en la carpeta assets 145x 45px
            ],className='mb-2'), # mb-2 permite crear espacio : https://hackerthemes.com/bootstrap-cheatsheet/
            dbc.Card([
                    dbc.CardBody([
                        dbc.CardLink("Profile David", target="_blank",
                                    href="https://www.linkedin.com/in/david-bustos-usta-180676a0/")
                    ])
            ]),
        ],# Cerrar columna 1 fila 1
            width=2), # Tamaño de la columna 2 pixeles
        dbc.Col([ # Columna 2 fila 1
            dbc.Card([
                dbc.CardBody([ # Estos son cards vacios 
                    dcc.DatePickerSingle(   # https://dash.plotly.com/dash-core-components
                    id='my-date-picker-start',
                    date=date(2020,1,1), # Fecha elegida por defecto
                    className='ml-5' # css class ml-5 margin left 5 entre el card y el picker
                    ),
                    dcc.DatePickerSingle(
                    id='my-date-picker-end',
                    date=date(2021,4,4), # fecha por defecto
                    className='mb-2 ml-2' # margin botton 2 and margin left 2
                    ),
                ])
            ], color="info"),
        ]# Cerrar fila 1 columna 2
            , width=8), # Tamaño mucho mayor de columna 2 en este caso 8 pixeles 
    ],# Cerrar primera fila (1)
        className='mb-2 mt-2'), # espacio de bottom 2 y top 2 units
    
    dbc.Row([ # Abrir segunda fila (2)
        dbc.Col([
            dbc.Card([  # https://lottiefiles.com/
                    dbc.CardHeader(Lottie(options=options,width="25%",height="25%",url=url_coonections)),
                    dbc.CardBody([
                        html.H6('Conexiones'),
                        html.H2(id='content-connections',children="000")
                    ],style={'textAlign':'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options,width="32%",height="32%",url=url_companies)),
                dbc.CardBody([
                    html.H6('Compañias'),
                    html.H2(id='content-companies',children="000")
                ],style={'textAlign':'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options,width="25%",height="25%",url=url_msg_in)),
                dbc.CardBody([
                    html.H6('Invitaciones recibidas'),
                    html.H2(id='content-msg-in',children ="000")
                ],style={'textAlign':'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options,width="53%",height="53%",url=url_msg_out)),
                dbc.CardBody([
                    html.H6('Invitaciones enviadas'), # dash html component tamaño 6
                    html.H2(id='content-msg-out',children="000") # dash html tamaño 2
                ],style={'textAlign':'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options,width="25%",height="25%",url=url_reactions)),
                dbc.CardBody([
                    html.H6('Reactions'),
                    html.H2(id='content-reactions',children='000')
                ],style={'textAlign':'center'})
            ]),
        ], width=2),
    ], # Cerrar segunda fila (2)
        className='mb-2'),
    dbc.Row([ # Abrir tercera fila (3)
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='line-chart',figure={},config={'displayModeBar':False}),
                ])
            ]),
        ], width=6),
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='bar-chart',figure={}),
                ])
            ]),
        ], width=4),
    ],# Cerrar 3 fila (3)
        className='mb-2'),
    dbc.Row([ # Abri cuarta fila (4)
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='TBD',figure={}),
                ])
            ]),
        ], width=3),
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='pie-chart',figure={}),
                ])
            ]),
        ], width=3),
        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='wordcloud',figure={}),
                ])
            ]),
        ], width=4),
    ],# Cerrar cuarta fila (4)
        className='mb-2'),
],# Cerrar el container
    fluid=True) # Hay una diferencia clara entre un fluid TRUE y FALSE en la forma como se ve el dashboard


# Parte del callback 
@app.callback(
    Output('content-connections','children'), # estos son los 5 outputs que seran texto
    Output('content-companies','children'),
    Output('content-msg-in','children'),
    Output('content-msg-out','children'),
    Output('content-reactions','children'),
    Input('my-date-picker-start','date'), # Estos son los inputs de las fechas
    Input('my-date-picker-end','date')
)

def update_small_cards(start_date, end_date):
    # Conexiones
    dff_c = df_cnt.copy()
    dff_c = dff_c[(dff_c['Connected On']>=start_date) & (dff_c['Connected On']<=end_date)]
    conctns_num = len(dff_c)
    compns_num = len(dff_c['Company'].unique())

    # Invitaciones
    dff_i = df_invite.copy()
    dff_i = dff_i[(dff_i['Sent At']>=start_date) & (dff_i['Sent At']<=end_date)]
    # print(dff_i)
    in_num = len(dff_i[dff_i['Direction'] == 'INCOMING'])
    out_num = len(dff_i[dff_i['Direction'] == 'OUTGOING'])

    # Reaacciones
    dff_r = df_react.copy()
    dff_r = dff_r[(dff_r['Date']>=start_date) & (dff_r['Date']<=end_date)]
    reactns_num = len(dff_r)

    return conctns_num, compns_num, in_num, out_num, reactns_num


app.run_server(mode='external', port = 8003, dev_tools_ui=True, #debug=True,
              dev_tools_hot_reload =True, threaded=True)
```
