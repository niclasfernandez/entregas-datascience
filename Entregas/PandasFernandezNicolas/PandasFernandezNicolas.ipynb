Ejercicio 2 --> Separar nombres de apellidos en dos columnas distintas usando la coma como separador

```python
data[['First Name','Last Name']] = data.Athlete.str.split(", ",1,expand=True)
data
```

Ejercicio 3 --> Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?

```python
data.groupby(['Country_Code']).size().sort_values(ascending=False).to_frame('Medals').iloc[0,:]
```
