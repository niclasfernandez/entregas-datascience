```python
df.drop(['duration_ms', 'mode', 'time_signature', 'explicit', 'key', 'mode'], axis=1, inplace=True)
df.describe().T
```

```python
# Categorizando popularity
def f(row):
    if row['popularity'] < 65:
        val = "low"
    else:
        val = "high"
    return val

df['popularityCategorical'] = df.apply(f, axis=1)
df.drop(['popularity'], axis=1, inplace=True)
df.head()
```

```python
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.pairplot(df)
    plt.show()
```

```python
#Correlaciones con df2 (truncated)
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    plt.figure(dpi = 120,figsize= (5,4))
    mask = np.triu(np.ones_like(df2.corr(),dtype = bool))
    sns.heatmap(df2.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
    plt.yticks(rotation = 0)
    plt.xticks(rotation = 90)
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
#JoinPlot de danceability con todas las otras
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.color_palette("Set2")
    plt.figure(dpi = 100, figsize = (5,4))
    print("Joint plot de danceability con otras variables ==> \n")
    for i in  df2.columns:
        if i != 'danceability' and i != 'id' and i != 'name' and i != 'artists' and i != 'id_artists' and i != 'release_date' and i != 'popularityCategorical':
            print(f"Correlacion entre danceability y {i} ==> ",df2.corr().loc['danceability'][i])
            g = sns.jointplot(x='danceability',y=i,data=df2,kind = 'reg', scatter = False)
            #g.ax_joint.scatter(df2.danceability,i, c='purple')
            plt.show()
```

```python
#Correlaciones con df (All info)
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    plt.figure(dpi = 120,figsize= (5,4))
    mask = np.triu(np.ones_like(df.corr(),dtype = bool))
    sns.heatmap(df.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
    plt.yticks(rotation = 0)
    plt.xticks(rotation = 90)
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
#JoinPlot de popularity vs las otras
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.color_palette("Set2")
    plt.figure(dpi = 100, figsize = (5,4))
    print("Joint plot de popularity con otras variables ==> \n")
    for i in  df.columns:
        if i != 'popularity' and i != 'id' and i != 'name' and i != 'artists' and i != 'id_artists' and i != 'release_date' and i != 'popularityCategorical':
            print(f"Correlacion entre popularity y {i} ==> ",df.corr().loc['popularity'][i])
            g = sns.jointplot(x='popularity',y=i,data=df,kind = 'reg', scatter = False)
            plt.show()
```
