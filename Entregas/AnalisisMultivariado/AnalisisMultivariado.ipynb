```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../datasetsGrandes/spotifySongs/tracks.csv')
df.head()
```

```python
df.drop(['duration_ms', 'mode', 'time_signature', 'explicit', 'key', 'mode'], axis=1, inplace=True)
df.describe().T
```

```python
def f(row):
    if row['popularity'] < 65:
        val = "low"
    else:
        val = "high"
    return val

df['popularityCategorical'] = df.apply(f, axis=1)
df.drop(['popularity'], axis=1, inplace=True)
df.head()
```

```python
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.pairplot(df)
    plt.show()
```

```python
with plt.rc_context({'axes.edgecolor':'orange', 'xtick.color':'red', 'ytick.color':'green', 'figure.facecolor':'white'}):
    sns.pairplot(df, hue = 'popularityCategorical')
    plt.show()
```
