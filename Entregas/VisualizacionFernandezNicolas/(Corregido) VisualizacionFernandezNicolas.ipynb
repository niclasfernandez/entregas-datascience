Rango de salario más popular --> 65-75K aprox

```python
#Para que se pueda ver mejor podés aplicar el subplots() antes de llamar el ax
fig,ax = plt.subplots(figsize=(12,6))
ax = df.Salary.plot(kind="hist", bins=50, figsize=(12,6), color="green")
ax.set_xlabel('$')
ax.set_ylabel('Amount of People')
ax.legend()
ax.set_title('Salary Histogram')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

#Figsize en el ax no es lo mismo que en el fig. Si el fig es más chico que el ax te va a quedar embebido
#fig y ax son el unpacking de una tupla que devuelve subplots. Si bien los nombres son convencion, 
#el fig refiere al marco donde se va a insertar el ax
#el ax contiene todo lo relacionado al gráfico en si

#Mas info
#https://towardsdatascience.com/clearing-the-confusion-once-and-for-all-fig-ax-plt-subplots-b122bb7783ca
```

```python
#Mas allá de eso, no te parece mucho 50 bins? Como deducirias el rango de salarios con tanta dispersion?
#A ver cómo quedaría asi
rango= list(range((df.Salary.min()//1000)*1000,df.Salary.max(), 20000)) #//1000*1000 es para truncarlo al millar
fig,ax = plt.subplots(figsize=(12,6))
ax = df.Salary.plot(kind="hist", bins=rango, figsize=(12,6), color="green")
ax.set_xlabel('$')
ax.set_ylabel('Amount of People')
ax.legend()
ax.set_title('Salary Histogram')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

ax.set_xticks(rango) #Y lo ponemos en el eje x
plt.show() #Esto hace que no se muestren todos los objetos intermedios. Solo el grafico
```

```python
df.loc[df['GenderID'] == 1, 'Gender'] = "Women"
df.loc[df['GenderID'] == 0, 'Gender'] = "Men" #Ojo que esto está al revés!
ax = sns.catplot(data=df, kind='violin', x='Gender', y='Salary')
ax.set(xlabel='Genero', ylabel='Distribución de salarios', title='Categorización de la distribución de salarios')
```

```python
#Fijate, si repetimos con Sex 
ax = sns.catplot(data=df, kind='violin', x='Sex', y='Salary')
ax.set(xlabel='Genero', ylabel='Distribución de salarios', title='Categorización de la distribución de salarios')
plt.show()
```

```python
#En este caso, estaria bueno quizás verlo así
fig,ax= plt.subplots(nrows=1,ncols=2, figsize=(12,6))
fig.tight_layout()

ax[0].sharey(ax[1])
ax[0].plot(df.groupby(df.DateofHire.dt.year).DateofHire.count())
ax[1].plot(df.groupby(df.DateofHire.dt.quarter).DateofHire.count())

ax[0].set_title('Hirings by year')
ax[1].set_title('Hirings by quarter')
ax[1].set_xticks(list(range(1,5)))

plt.show()
#El resto de la complejización te lo dejo a vos!
```
