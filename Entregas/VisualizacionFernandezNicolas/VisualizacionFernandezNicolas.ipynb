```python
# Imports
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# 2 decimals for everything
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("./HRDataset_v14.csv")
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df.head()
```

```python
ax = df.Salary.plot(kind="hist", bins=50, figsize=(12,6), color="green")
ax.set_xlabel('$')
ax.set_ylabel('Amount of People')
ax.legend()
ax.set_title('Salary Histogram')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
```

```python
df.loc[df['GenderID'] == 1, 'Gender'] = "Women"
df.loc[df['GenderID'] == 0, 'Gender'] = "Men"
ax = sns.catplot(data=df, kind='violin', x='Gender', y='Salary')
ax.set(xlabel='Genero', ylabel='DistribuciÃ³n de salarios', title='CategorizaciÃ³n de la distribuciÃ³n de salarios')
```
