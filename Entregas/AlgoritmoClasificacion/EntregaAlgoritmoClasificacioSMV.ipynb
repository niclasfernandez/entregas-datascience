```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np

# Preprocesado y modelado
# ==============================================================================
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC, SVR
from mlxtend.plotting import plot_decision_regions

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading dataset
df = pd.read_csv('../../datasetsGrandes/spotifySongs/tracks.csv')
df.count()
```

```python
def categorize(row):
    if row['popularity'] < 70:
        val = "low"
    else:
        val = "high"
    return val

df_short = df.drop(df[(df.popularity > 30) & (df.popularity < 70)].index)
df['popularityCategorical'] = df.apply(categorize, axis=1)
df.drop(['popularity'], axis=1, inplace=True)
df_short['popularityCategorical'] = df_short.apply(categorize, axis=1)
df_short.drop(['popularity'], axis=1, inplace=True)
df.head()
```

```python
#Separamos en X e y
X = df.drop(['duration_ms', 'mode', 'time_signature', 'explicit', 'key', 'mode', 'id', 'name', 'artists', 'id_artists', 'release_date', 'popularityCategorical'], axis=1)
y = df.popularityCategorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
estimador_svm = SVC()
estimador_svm.fit(X_train, y_train)
```
